```python
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
categorical_cols = [
    'workclass', 'marital-status', 'occupation',
    'relationship', 'race', 'gender',
    'native-country', 'education'
]
```

```python
encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    encoders[col] = le
```

```python
joblib.dump(encoders, "label_encoders.pkl")
```

```python
x = data.drop('income', axis=1)
y = data['income']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# 📌 7. Scale features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# 📌 8. Define models
models = {
    'Logistic Regression': LogisticRegression(max_iter=500),
    'Random Forest': RandomForestClassifier(),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC(),
    'Neural Network': MLPClassifier(max_iter=500)
}

# 📌 9. Train, predict, score
result = {}

for name, model in models.items():
    model.fit(x_train, y_train)
    preds = model.predict(x_test)
    acc = accuracy_score(y_test, preds)
    result[name] = acc

print("✅ Results:", result)
```

```python
# 📌 10. Plot
plt.figure(figsize=(10, 6))
plt.bar(result.keys(), result.values(), color='skyblue')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import pandas as pd
import joblib

# ✅ Save the feature names before converting to NumPy
feature_columns = [
    'age', 'workclass', 'education', 'marital-status', 'occupation',
    'relationship', 'race', 'gender', 'native-country',
    'hours-per-week', 'capital-gain', 'capital-loss', 'educational-num', 'fnlwgt'
]
joblib.dump(feature_columns, "feature_order.pkl")
# ✅ Save encoders
joblib.dump(encoders, "label_encoders.pkl")

# ✅ Save scaler
joblib.dump(scaler, "scaler.pkl")

# ✅ Save all models together
joblib.dump(models, "all_models.pkl")

# ✅ Save each model individually
for name, model in models.items():
    filename = f"{name.replace(' ', '_').lower()}_model.pkl"
    joblib.dump(model, filename)

print("✅ All files saved: feature_order, encoders, scaler, models ✅")
```
